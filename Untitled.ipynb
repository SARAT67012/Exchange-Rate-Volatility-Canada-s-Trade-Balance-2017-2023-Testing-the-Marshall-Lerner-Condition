```python
# STEP 1: Import libraries and load raw CSV files

import pandas as pd
import numpy as np

# 1.1 – Read Canada trade data (annual)
trade = pd.read_csv("canada_trade_2000_2023.csv")

# 1.2 – Read daily CAD/USD exchange rate
cad_usd = pd.read_csv("cad_usd_rate.csv")

# 1.3 – (Optional) Read CAD/EUR data if needed later
cad_eur = pd.read_csv("cad_eur.csv")

# 1.4 – Quickly check first few rows to confirm files loaded correctly
print("Trade data:")
print(trade.head(), "\n")

print("CAD/USD data:")
print(cad_usd.head(), "\n")

print("CAD/EUR data:")
print(cad_eur.head())
```

```python
# STEP 2: Clean daily CAD/USD rate and convert to annual averages

# 2.1 – Convert the 'date' column to datetime format
cad_usd['date'] = pd.to_datetime(cad_usd['date'])

# 2.2 – Extract year from the date
cad_usd['Year'] = cad_usd['date'].dt.year

# 2.3 – Compute the annual average exchange rate
annual_er = cad_usd.groupby('Year')['value'].mean().reset_index()

# 2.4 – Rename the column for clarity
annual_er = annual_er.rename(columns={'value': 'CAD_USD_avg'})

# 2.5 – Display the annual exchange rate table
print(annual_er.head(10))
```

```python
# STEP 3: Merge annual ER with trade data

# 3.1 – Ensure 'Year' in trade dataset is integer
trade['Year'] = trade['Year'].astype(int)

# 3.2 – Merge trade data with annual average exchange rate
df = pd.merge(trade, annual_er, on='Year', how='inner')

# 3.3 – Display first few rows of merged dataset
print(df.head(10))

# 3.4 – Check if any years are missing after merge
print("Years in merged data:", df['Year'].min(), "to", df['Year'].max())
print("Total rows:", len(df))
```

```python
# STEP 4: Prepare variables for regression

# 4.1 – Check exchange rate direction (important!)
# If CAD_USD_avg is given as USD per 1 CAD (e.g., 0.75), convert to CAD per USD
df['ER_CAD_per_USD'] = 1 / df['CAD_USD_avg']

# 4.2 – Replace 0 values in exports/imports if any (log cannot take zero)
df['Exports_Billion_USD'] = df['Exports_Billion_USD'].replace(0, np.nan)
df['Imports_Billion_USD'] = df['Imports_Billion_USD'].replace(0, np.nan)

# 4.3 – Create log variables
df['ln_exports'] = np.log(df['Exports_Billion_USD'])
df['ln_imports'] = np.log(df['Imports_Billion_USD'])
df['ln_ER'] = np.log(df['ER_CAD_per_USD'])

# 4.4 – Optional: Create trade balance ratio (Exports / Imports)
df['TB_ratio'] = df['Exports_Billion_USD'] / df['Imports_Billion_USD']
df['ln_TB_ratio'] = np.log(df['TB_ratio'])

# 4.5 – Display final clean dataset
print(df[['Year','Exports_Billion_USD','Imports_Billion_USD','ER_CAD_per_USD','ln_exports','ln_imports','ln_ER','ln_TB_ratio']].head())
```

```python
# STEP 5: Export Elasticity Regression

import statsmodels.api as sm

# 5.1 – Define dependent and independent variables
Y = df['ln_exports']
X = df[['ln_ER']]
X = sm.add_constant(X)   # adds intercept term

# 5.2 – Run OLS regression
model_exports = sm.OLS(Y, X).fit()

# 5.3 – Print summary
print(model_exports.summary())
```

```python
# STEP 6: Import Elasticity Regression

# 6.1 – Define dependent and independent variables
Y = df['ln_imports']
X = df[['ln_ER']]
X = sm.add_constant(X)

# 6.2 – Run OLS regression
model_imports = sm.OLS(Y, X).fit()

# 6.3 – Show results
print(model_imports.summary())
```

```python
# STEP 7: Marshall-Lerner Condition

alpha_1 = model_exports.params['ln_ER']
beta_1 = model_imports.params['ln_ER']

ML_sum = abs(alpha_1) + abs(beta_1)

print("Export elasticity (alpha1):", alpha_1)
print("Import elasticity (beta1):", beta_1)
print("Sum of elasticities:", ML_sum)

if ML_sum > 1:
    print("Marshall–Lerner condition HOLDS.")
else:
    print("Marshall–Lerner condition DOES NOT hold.")
```

```python
import numpy as np

plt.figure(figsize=(6, 5))
x = df['ln_ER']
y = df['ln_exports']

# Scatter
plt.scatter(x, y, label='Data points')

# Regression line using numpy
b1, b0 = np.polyfit(x, y, 1)   # slope, intercept
x_line = np.linspace(x.min(), x.max(), 100)
y_line = b1 * x_line + b0
plt.plot(x_line, y_line, label=f'Fit: ln_exports = {b0:.2f} + {b1:.2f} ln_ER')

plt.xlabel('ln(ER_CAD_per_USD)')
plt.ylabel('ln(Exports)')
plt.title('Export Elasticity: ln(Exports) vs ln(ER)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
